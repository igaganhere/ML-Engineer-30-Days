```python
from sklearn.datasets import fetch_california_housing
```

```python
df = fetch_california_housing(as_frame=True)
```

```python
df
```

```python
x = df.data.head(2000).copy()
```

```python
y = df.target.head(2000).copy()
```

```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
x['Noise_Uniform'] = np.random.rand(len(x))
x['Noise_Normal'] = np.random.randn(len(x))
x['Noise_Integers'] = np.random.randint(0, 100, len(x))
```

```python
from sklearn.feature_selection import mutual_info_regression as MIR
```

```python
mir_score = pd.Series(MIR(x,y,random_state=42), index=x.columns).sort_values(ascending=False)
```

```python
print(f'Mutual info score:\n {mir_score}')
```

```python
keep = mir_score[mir_score>0.01].index.tolist()
```

```python
print(f'Values kept after MIR score:\n{keep}')
```

```python
from sklearn.preprocessing import StandardScaler as SS
```

```python
x_scaled = SS().fit_transform(x[keep])
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
x_pca=pca.fit_transform(x_scaled)
```

```python
df_pca=pd.DataFrame(x_pca, columns=['PC1','PC2'])
```

```python
df_pca
```

```python
var1,var2=pca.explained_variance_ratio_
```

```python
print(f"PC1: {var1*100:.2f}% | PC2: {var2*100:.2f}% | Total: {(var1+var2)*100:.2f}%")
```

```python
#using 90% method
```

```python
pca=PCA(n_components=0.90)
```

```python
X_pca = pca.fit_transform(x_scaled)

num_components_kept = X_pca.shape[1]
print(f"--> PCA dynamically retained {num_components_kept} components to hit the 90% threshold.")

component_names = [f'PC{i+1}' for i in range(num_components_kept)]

df_pca = pd.DataFrame(data=X_pca, columns=component_names)
```
