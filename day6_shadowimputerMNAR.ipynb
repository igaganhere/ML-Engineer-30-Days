```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
age=np.random.uniform(25,70,12).round(0)
```

```python
true_income=((age*1200)+np.random.normal(0,15000,12)).round(0)
```

```python
df=pd.DataFrame({'Age':age,'Income':true_income})
```

```python
df
```

```python
df.loc[df.Income>85000,'Income'] =np.nan
```

```python
df
```

```python
#Original data set with MNAR problem 
```

```python
#Applying shadow imputer method 
```

```python
from sklearn.impute import SimpleImputer
```

```python
simple_imputer =SimpleImputer(strategy='mean', add_indicator=True)
```

```python
imputed_arr=simple_imputer.fit_transform(df)
```

```python
df_shadow=pd.DataFrame(imputed_arr, columns=['Age','Income','Missing_income_flag'])
```
